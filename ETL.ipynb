```python
import pandas as pd
import sqlite3
```

```python
csv_path = "Resources/god_data.csv"
gods_data = pd.read_csv(csv_path)
gods_data
```

```python
conn = sqlite3.connect('Resources/god_data.sqlite')
create_sql = "CREATE TABLE IF NOT EXISTS god_data (god_name TEXT, description TEXT, mythology TEXT, godly_aspects TEXT, godly_aspects2 TEXT, Img_URL TEXT)"
cursor = conn.cursor()
cursor.execute(create_sql)
gods_data.to_sql("god_data", con = conn, if_exists='replace')
```
